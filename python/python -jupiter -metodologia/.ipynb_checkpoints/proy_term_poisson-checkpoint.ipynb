```python

import numpy as np 
añostr = 'C:/Users/ASTRONAUTA/Documents/UNADM PROYECTO TERMINAL/datos enfermedades/ssa_egresos_2012/EGRESO_2012.csv'
def enfdb(año,edi,edf):
    enf12 = pd.read_csv(año)
    enf12=enf12[['INGRE','EDAD','SEXO','MUNIC','AFECPRIN']]


    for a in range(enf12.shape[0]):
        if isinstance(enf12.loc[a,'AFECPRIN'],str) == False :
            print(a,enf12.loc[a,'MUNIC'])
            enf12=enf12.drop(a, axis=0)
            
    enf12['EDAD']=enf12['EDAD'].astype(float)
    enf12['MUNIC']=enf12['MUNIC'].astype(float)
    enf12['AFECPRIN']=enf12['AFECPRIN'].astype(str)
    enf12=enf12.loc[enf12['MUNIC'] == 39]
    enf12=enf12[enf12["AFECPRIN"].str.contains("J")]
    enf12=enf12.loc[(enf12['EDAD'] >= edi) & (enf12['EDAD'] <= edf )]
    enf12=enf12.reset_index()
    
    enf12['INGRE'] = pd.to_datetime(enf12['INGRE'])
    enf12=enf12.sort_values(by='INGRE',ignore_index=True)
    
    time = pd.to_datetime('2012/01/01')
    aft = pd.to_datetime('2012/12/31')
    enf12=enf12.loc[(enf12['INGRE'] >= time)&((enf12['INGRE'] <= aft))]
    
    enf12=enf12.reset_index()
    
    date12 = pd.read_csv('C:/Users/ASTRONAUTA/Documents/UNADM PROYECTO TERMINAL/datos enfermedades/dates.csv')
    date12['date'] = pd.to_datetime(date12['date'])
    date12=date12.sort_values(by='date',ignore_index=True)
    #date12=date12['date'].tolist()
    
    ling=enf12['INGRE'].tolist()
    p=1
    h = 0
    lenf=[]
    
    for g in range(enf12.shape[0]): 
        if g == enf12.shape[0]-1:
            if enf12.loc[g,'INGRE']!=enf12.loc[g-1,'INGRE']:
                lenf.append(1)
            continue       
        
        elif date12.loc[h,'date'] == enf12.loc[g,'INGRE']:
            p+=1
            if enf12.loc[g,'INGRE'] != enf12.loc[g+1,'INGRE']:
                h=h+1
                lenf.append(p)
                p=1
                while enf12.loc[g+1,'INGRE'] != date12.loc[h,'date']:
                    h=h+1
                    lenf.append(0)
                                       
        else:
            h+=1

    
    print(len(lenf))               

 
    j=0
    for i in lenf:
        j = j+i
    media = j/len(lenf)
    print(media)
        
    if len(lenf)<len(date12):
        dif = len(date12)-len(lenf)
        for i in range(dif):
            lenf.append(media)
    #print(len(lenf))
    #lenf = np.array(lenf)
    return lenf
```
